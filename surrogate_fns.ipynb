```python
def step(x):
    return x>=0
def hinge(x):
    return np.maximum(0, 1+ x)
def softplus(x):
    return np.log(1+np.exp(x)) / np.log(2)

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-2, 2, 1000)
plt.close()
plt.plot(x, step(x), label='Step-indicator', lw=3)
plt.plot(x, hinge(x), label='Hinge', lw=3)
plt.plot(x, softplus(x), label='Softplus', lw=3)
plt.plot(x, sigmoid(x), label='Sigmoid', lw=3)
plt.legend(fontsize=20)
sns.despine()
plt.xlabel(r'$z$', fontsize=20)
plt.ylabel(r'$h(z)$', fontsize=20)
plt.savefig('./figures/surrogates.png', dpi=300)
plt.show()
```
